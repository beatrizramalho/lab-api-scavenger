# Challenge 2: Count Commits

```python
#1. Obtain all the commits made in the past 2 weeks via API, which is a JSON array that contains multiple commit objects.

response2 = requests.get('https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/forks')
response2.json()
```

```python
df2 = pd.DataFrame(response2.json())
df2.head(50)
```

```python
#to know the name of the columns:
df.info()
```

```python
#checking the column pushed_at to know the format of the dates:
df2['pushed_at']
```

```python
#Convert into a format date that can be worked
import datetime 
df2['pushed_at'] = pd.to_datetime(df2['pushed_at'], format= "%Y-%m-%dT%H:%M:%SZ")
```

```python
#Checking the new data format:
df2['pushed_at']
```

```python
#Filter for last two weeks of december
df2 = df2[df2['pushed_at'] > "2021-12-19 00:00:00"]
df2
```

```python
#Count how many commit objects are contained in the array.
print("The number of commits done in the last two weeks of 2021 is", len(df2.index))
```

```python
#If we want only for the last two weeks from today:
week_ago2 = datetime.datetime.now() - datetime.timedelta(days=14)
week_ago2

df3 = df2[df2['pushed_at'] > week_ago2]
df3
```

```python
print("The number of commits done in the last two weeks is", len(df3.index))
```
